```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
from imutils.video import VideoStream
import numpy as np
import argparse
import imutils
import time
import dlib
import os
Phase='ph1'
```

```
import pandas as pd
df = pd.DataFrame(columns=[Phase+'Frameid',Phase+'bb',Phase+'Path_FrameId',Phase+'BB_Coordinates',Phase+'Path_to_BB_File',Phase+'CropFilePath'])
len(df)
```

```
os.chdir('/content/drive/My Drive/Lennox/')
detector = dlib.simple_object_detector("/content/drive/My Drive/Lennox/Phase1Coil.dat")
imagesList=os.listdir('/content/drive/My Drive/Lennox/Phase1/Images')
imagesList
```

```
for i in imagesList:
  # filename='/content/drive/My Drive/Lennox/Images/'+i
  filename='/content/drive/My Drive/Lennox/Phase1/Images/'+i
  BBfilename='/content/drive/My Drive/Lennox/Phase1/Images_BB/'+i
  Resultfilename='/content/drive/My Drive/Lennox/Phase1/crop/'+i
  image = cv2.imread(filename)
  print('Original Dimensions : ',image.shape)
 
  scale_percent = 95 # percent of original size
  width = int(image.shape[1] * scale_percent / 100)
  height = int(image.shape[0] * scale_percent / 100)
  dim = (width, height)
  # resize image
  resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
  print('Resized Dimensions : ',resized.shape)

  imgcopy = np.zeros(resized.shape[:3], np.uint8)##
  #imgcopy = np.zeros(image.shape[:3], np.uint8)##
  np.copyto(imgcopy, resized)## 

  detections = detector(resized)

  for k, d in enumerate(detections):
      print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(
            k, d.left(), d.top(), d.right(), d.bottom()))
      x,y,w,h = d.left(), d.top(), d.right(), d.bottom()
      print('Dlib Detections-',x,y,w,h)
      # cv2.rectangle(resized, (x, y), (w,h), [0,255,255], 2)
      # cv2_imshow(resized)
      # crop=resized[y:h,x:w]
      # cv2_imshow(crop)
      # CroppedFileName=Resultfilename.split(".")[0]+Phase+'.jpg'
      # print(CroppedFileName)     
      # cv2.imwrite(CroppedFileName,crop)  
      # crop=resized[y:h,x:w]
      # cv2_imshow(crop)
      # CroppedFileName=Resultfilename.split(".")[0]+'_'+str(k)+'_'+Phase+'.jpg'
      # # print(CroppedFileName)
      # cv2.imwrite(CroppedFileName,crop)

      # cv2.rectangle(resized, (x, y), (w,h), [0,255,255], 2)
      # # cv2_imshow(resized)      
      # cv2.imwrite(BBfilename.split(".")[0]+'_'+Phase+'bb.jpg',resized)
      if x > 0  and y> 0 and w > 0 and h >0:
        crop=resized[y:h,x:w]
        cv2_imshow(crop)
        CroppedFileName=Resultfilename.split(".")[0]+'_'+str(k)+'_'+Phase+'.jpg'
        print('crop file',CroppedFileName)
        print('crop size',crop.shape)
        cv2.imwrite(CroppedFileName,crop)

        imgcopy1 = np.zeros(resized.shape[:3], np.uint8)##
        np.copyto(imgcopy1, imgcopy)##
        cv2.rectangle(imgcopy1, (x, y), (w,h), [0,255,255], 2)##
        # cv2_imshow(imgcopy1)
        InterimBBImg=BBfilename.split(".")[0]+'_'+str(k)+'_'+Phase+'bb.jpg'
        cv2.imwrite(InterimBBImg,imgcopy1) ##
        print('InterimBBImg',InterimBBImg)

        cv2.rectangle(resized, (x, y), (w,h), [0,255,255], 2)
        # cv2_imshow(resized)
        FinalBBImg=BBfilename.split(".")[0]+'_'+Phase+'bb.jpg' ##     
        cv2.imwrite(FinalBBImg,resized)##
        print('FinalBBImg',FinalBBImg)     
      
        # df.loc[len(df)]= [Phase,i.split(".")[0],filename.split("Lennox/")[1],[x,y,w,h],BBfilename.split("Lennox/")[1],Resultfilename.split("Lennox/")[1]] 
        df.loc[len(df)]= [i.split(".")[0],i.split(".")[0]+'_'+str(k)+'_'+Phase,filename.split("Lennox/")[1],[x,y,w,h],InterimBBImg.split("Lennox/")[1],CroppedFileName.split("Lennox/")[1]] ##
      else:
        print('do nothing')
    
      #cv2.imwrite('testing'+str(count+1)+'.png',image)
```
